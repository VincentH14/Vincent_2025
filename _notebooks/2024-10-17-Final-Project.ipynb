```python

students = {
    'Alice': 85,
    'Bob': 72,
    'Charlie': 90,
    'Diana': 60,
    'Ethan': 78,
    'Fiona': 94,
    'George': 88,
    'Hannah': 95,
    'Ian': 67,
    'Jasmine': 82
}

def calculate_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'


def calculate_average(students):
    total_score = 0
    count = 0
    for score in students.values():
        total_score += score
        count += 1
    return total_score / count if count > 0 else 0


def print_student_grades(students):
    for student, score in students.items():
        grade = calculate_grade(score)
        
        
        output_message = f"{student} scored {score} and received a grade of {grade}."
        print(output_message)


def add_student(students):
    name = input("Enter the student's name: ")
    while True:
        try:
            score = float(input(f"Enter {name}'s score (0-100): "))
            if 0 <= score <= 100:
                students[name] = score
                print(f"{name} has been added with a score of {score}.")
                break
            else:
                print("Please enter a valid score between 0 and 100.")
        except ValueError:
            print("Invalid input! Please enter a numeric value.")

if __name__ == "__main__":
    print("Student Grades:")
    print_student_grades(students)
 
    average_score = calculate_average(students)
    print(f"\nThe average score of the students is: {average_score:.2f}")
   
    add_more = input("\nWould you like to add a new student? (yes/no): ").strip().lower()
    while add_more == 'yes':
        add_student(students)
        print_student_grades(students)
        average_score = calculate_average(students)
        print(f"\nThe updated average score of the students is: {average_score:.2f}")
        add_more = input("\nWould you like to add another student? (yes/no): ").strip().lower()
    
    print("Thank you for using the Student Grade System!")
```

Variables:

The students dictionary stores student names as keys and their corresponding scores as values.


Data Abstraction:

The calculate_grade function includes the logic for determining a letter grade based on a student's score. This removes the complexity from the program and makes it easier and cleaner to read and understand


Strings:

The output messages that display each student's score and grade are constructed using formatted strings. This involves strings to create informative messages.


Conditionals:

The if, elif, else functions are used within the calculate_grade function checks different score ranges to give you the right grade.This structure allows the program to output various outcomes based on the input score.


Iterations:

The for loop iterates through the students dictionary to process each student's score, allowing the program to handle multiple students. Another iteration is used in the calculate_average function to add all the scores.
