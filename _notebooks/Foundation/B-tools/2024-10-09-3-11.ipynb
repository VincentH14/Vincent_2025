```python

<style>
        body { font-family: Arial, sans-serif; background-color: #f0f0f0; padding: 20px; }
        .npc { font-weight: bold; }
        .question { margin: 10px 0; }
        .response { margin-left: 20px; }
</style>


<script>
    // 2D array for NPC dialogues and questions
    const npcs = [
        ["Alice", "Hello! Do you want to answer a question?", "What is your favorite color?", ["Red", "Blue", "Green"]],
        ["Bob", "Hi there! I have a riddle for you.", "What has keys but can't open locks?", ["Piano", "Map", "Door"]],
        ["Charlie", "Greetings! Would you like to play a game?", "What is 2 + 2?", ["3", "4", "5"]]
    ];

    // Function to display NPC dialogues and handle player responses
    function displayGame() {
        const gameDiv = document.getElementById("game");
        gameDiv.innerHTML = ""; // Clear previous content

        // Loop through each NPC
        for (let i = 0; i < npcs.length; i++) {
            const npcName = npcs[i][0];
            const npcDialogue = npcs[i][1];
            const question = npcs[i][2];
            const answers = npcs[i][3];

            // Display NPC name and dialogue
            gameDiv.innerHTML += `<div class="npc">${npcName}:</div>`;
            gameDiv.innerHTML += `<div class="response">${npcDialogue}</div>`;

            // Display question and possible answers
            gameDiv.innerHTML += `<div class="question">${question}</div>`;
            for (let j = 0; j < answers.length; j++) {
                gameDiv.innerHTML += `<button onclick="handleAnswer('${answers[j]}')">${answers[j]}</button> `;
            }

            gameDiv.innerHTML += "<hr>"; // Separator between NPCs
        }
    }

    // Function to handle player answers
    function handleAnswer(answer) {
        const gameDiv = document.getElementById("game");
        gameDiv.innerHTML += `<div class="response">You answered: ${answer}</div>`;
    }

    // Initialize the game
    displayGame();
</script>
```

```python
<style>
        canvas { border: 1px solid #000; }
</style>

<canvas id="spriteCanvas" width="400" height="400"></canvas>

<script>
    // Define sprite sheet image and metadata
    const spriteSheetImage = new Image();
    spriteSheetImage.src = 'path/to/sprite-sheet.png'; // Update with your sprite sheet path

    // Sprite metadata: [x, y, width, height]
    const sprites = [
        [0, 0, 32, 32],    // Sprite 1
        [32, 0, 32, 32],   // Sprite 2
        [64, 0, 32, 32],   // Sprite 3
        [96, 0, 32, 32],   // Sprite 4
        // Add more sprites as needed
    ];

    // Function to draw individual sprites
    function drawSprites() {
        const canvas = document.getElementById('spriteCanvas');
        const ctx = canvas.getContext('2d');

        // Loop through the sprites array and draw each sprite
        for (let i = 0; i < sprites.length; i++) {
            const [x, y, width, height] = sprites[i];
            ctx.drawImage(spriteSheetImage, x, y, width, height, i * 40, 0, width, height); // Display horizontally
        }
    }

    // Event to ensure the image is loaded before drawing
    spriteSheetImage.onload = function() {
        drawSprites();
    };
</script>
```
